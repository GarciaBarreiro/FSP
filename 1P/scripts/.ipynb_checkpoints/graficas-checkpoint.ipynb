```python
import csv
def read_csv(filename):   # returns dictionary of lists, good for working with graphs
    data = {'nodes': [], 'cpus': [], 'time': [], 'err': [], 'qual': [],'time_dt': [],'err_dt': [],'qual_dt': []}
    with open(filename, newline='') as csvfile:
        rdr = csv.reader(csvfile, delimiter=',')
        try:
            row = rdr.__next__()
        except StopIteration as si:
            return None

        prev_cpu = int(row[1])
        prev_node = int(row[0])
        temp_list = {'nodes': [], 'cpus': [], 'time': [], 'err': [], 'qual': [],'time_dt': [],'err_dt': [],'qual_dt': []}
        
        continue_while = True
        while continue_while:
            if prev_cpu == int(row[1]) and prev_node == int(row[0]):
                temp_list['nodes'].append(int(row[0]))
                temp_list['cpus'].append(int(row[1]))
                temp_list['time'].append(float(row[2]))
                temp_list['err'].append(float(row[3]))
                temp_list['qual'].append(float(row[4]))
            else:
                data['nodes'].append(temp_list['nodes'][0])
                data['cpus'].append(temp_list['cpus'][0])
                data['time'].append(np.mean(temp_list['time']))
                data['err'].append(np.mean(temp_list['err']))
                data['qual'].append(np.mean(temp_list['qual']))
                data['time_dt'].append(np.std(temp_list['time']))
                data['err_dt'].append(np.std(temp_list['err']))
                data['qual_dt'].append(np.std(temp_list['qual']))

                temp_list = {'nodes': [], 'cpus': [], 'time': [], 'err': [], 'qual': [],'time_dt': [],'err_dt': [],'qual_dt': []}

            prev_cpu = int(row[1])
            prev_node = int(row[0])


            try:
                row = rdr.__next__()
            except StopIteration as si:
                continue_while = False

        # last iteration
        data['nodes'].append(temp_list['nodes'][0])
        data['cpus'].append(temp_list['cpus'][0])
        data['time'].append(np.mean(temp_list['time']))
        data['err'].append(np.mean(temp_list['err']))
        data['qual'].append(np.mean(temp_list['qual']))
        data['time_dt'].append(np.std(temp_list['time']))
        data['err_dt'].append(np.std(temp_list['err']))
        data['qual_dt'].append(np.std(temp_list['qual']))

        return data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import time

# Parmetros fijos
fichero = '../out_trapezoids/output-new.csv'
data = read_csv(filename=fichero)

# Crear DataFrames con los resultados
df_resultados = pd.DataFrame(data, columns=['nodes', 'cpus', 'time', 'err', 'qual','time_dt','err_dt','qual_dt'])

# Imprimir las tablas de resultados
print("\nTabla de Resultados:")
print(df_resultados.to_string(index=False))

#Calidad vs CPUS
plt.figure(figsize=(12, 10))
for cpu_count in df_resultados['cpus'].unique():
    subset = df_resultados[df_resultados['nodes'] == cpu_count]
    plt.plot(subset['cpus'], subset['qual'], label=f'Nodes={cpu_count}',marker='o')

plt.xlabel('CPUs')
plt.ylabel('Quality')
plt.title('Quality vs CPUs')
plt.legend()
plt.show()


#Calidad vs CPUS (Excluyendo CPUs = 1)
plt.figure(figsize=(12, 10))
for cpu_count in df_resultados['cpus'].unique():
    subset = df_resultados[(df_resultados['nodes'] == cpu_count) & (df_resultados['nodes'] != 1)]
    plt.plot(subset['cpus'], subset['qual'], label=f'Nodes={cpu_count}', marker='o')

plt.xlabel('CPUs')
plt.ylabel('Quality')
plt.title('Quality vs CPUs (Excluindo CPUs = 1)')
plt.legend()
plt.show()


#Error vs Nodes
plt.figure(figsize=(12, 10))
for cpu_count in df_resultados['cpus'].unique():
    subset = df_resultados[df_resultados['nodes'] == cpu_count]
    plt.plot(subset['cpus'], subset['err'],label=f'Nodes={cpu_count}',marker='o')

plt.xlabel('CPUS')
plt.ylabel('Error')
plt.title('Error vs CPUs')
plt.legend()
plt.show()

#Time vs Nodes
plt.figure(figsize=(10, 6))
for cpu_count in df_resultados['cpus'].unique():
    subset = df_resultados[df_resultados['nodes'] == cpu_count]
    plt.plot(subset['cpus'], subset['time'], label=f'Nodes={cpu_count}',marker='o')

plt.xlabel('CPUs')
plt.ylabel('Time')
plt.title('Variacin do Tempo segn Nodes e CPUs')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import time

# Parmetros fijos
file = '../out_montecarlo/output.csv'
data = read_csv(filename=file)

# Crear DataFrames con los resultados
df_resultados = pd.DataFrame(data, columns=['nodes', 'cpus', 'time', 'err', 'qual','time_dt','err_dt','qual_dt'])

# Imprimir las tablas de resultados
print("\nTabla de Resultados:")
print(df_resultados.to_string(index=False))

#Calidad vs CPUS
plt.figure(figsize=(10, 8))
for node_count in df_resultados['cpus'].unique():
    subset = df_resultados[df_resultados['nodes'] == node_count]
    plt.errorbar(subset['cpus'], subset['qual'], yerr=subset['qual_dt'], label=f'Nodes={node_count}', marker='o')

plt.xlabel('CPUs')
plt.ylabel('Quality')
plt.title('Quality vs CPUs (sen Desviacin Tpica)')
plt.legend()
plt.show()


#Calidad vs CPUS (Sen Desviacin Tpica)
plt.figure(figsize=(10, 8))
for node_count in df_resultados['cpus'].unique():
    subset = df_resultados[(df_resultados['nodes'] == node_count)]
    plt.errorbar(subset['cpus'], subset['qual'], label=f'Nodes={node_count}', marker='o')

plt.xlabel('CPUs')
plt.ylabel('Quality')
plt.title('Quality vs CPUs')
plt.legend()
plt.show()


#Error vs Nodes
plt.figure(figsize=(10, 8))
for node_count in df_resultados['cpus'].unique():
    subset = df_resultados[df_resultados['nodes'] == node_count]
    plt.errorbar(subset['cpus'], subset['err'], label=f'CPUs={node_count}', marker='o')

plt.xlabel('Nodes')
plt.ylabel('Error')
plt.title('Error vs CPUs')
plt.legend()
plt.show()

#Time vs Nodes
plt.figure(figsize=(10, 8))
for cpu_count in df_resultados['cpus'].unique():
    subset = df_resultados[df_resultados['nodes'] == cpu_count]
    plt.errorbar(subset['cpus'], subset['time'], label=f'Nodes={cpu_count}', marker='o')

plt.xlabel('CPUs')
plt.ylabel('Time')
plt.title('Variacin del Tiempo segn Nodes y CPUs')
plt.legend()
plt.show()
```

```python
import csv
def read_csv_2(filename):   # returns dictionary of lists, good for working with graphs
    data = {'nodes': [], 'cpus': [], 'time': [], 'err': [], 'qual': [],'time_dt': [],'err_dt': [],'qual_dt': [], 'iters': []}
    with open(filename, newline='') as csvfile:
        rdr = csv.reader(csvfile, delimiter=',')
        try:
            row = rdr.__next__()
        except StopIteration as si:
            return None

     
        prev_iters = 1
        temp_list = {'nodes': [], 'cpus': [], 'time': [], 'err': [], 'qual': [],'time_dt': [],'err_dt': [],'qual_dt': [], 'iters': []}
        
        continue_while = True
        while continue_while:
            if prev_iters == int(row[5]):
                temp_list['nodes'].append(4)
                temp_list['cpus'].append(16)
                temp_list['time'].append(float(row[2]))
                temp_list['err'].append(float(row[3]))
                temp_list['qual'].append(float(row[4]))
                temp_list['iters'].append(int(row[5]))
            else:
                data['nodes'].append(temp_list['nodes'][0])
                data['cpus'].append(temp_list['cpus'][0])
                data['time'].append(np.mean(temp_list['time']))
                data['err'].append(np.mean(temp_list['err']))
                data['qual'].append(np.mean(temp_list['qual']))
                data['iters'].append(temp_list['iters'][0])
                data['time_dt'].append(np.std(temp_list['time']))
                data['err_dt'].append(np.std(temp_list['err']))
                data['qual_dt'].append(np.std(temp_list['qual']))

                temp_list = {'nodes': [], 'cpus': [], 'time': [], 'err': [], 'qual': [],'time_dt': [],'err_dt': [],'qual_dt': [], 'iters': []}

            prev_iters = int(row[5])

            try:
                row = rdr.__next__()
            except StopIteration as si:
                continue_while = False

        # last iteration
        data['nodes'].append(temp_list['nodes'][0])
        data['cpus'].append(temp_list['cpus'][0])
        data['time'].append(np.mean(temp_list['time']))
        data['err'].append(np.mean(temp_list['err']))
        data['qual'].append(np.mean(temp_list['qual']))
        data['iters'].append(temp_list['iters'][0])
        data['time_dt'].append(np.std(temp_list['time']))
        data['err_dt'].append(np.std(temp_list['err']))
        data['qual_dt'].append(np.std(temp_list['qual']))

        return data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import time

# Parmetros fijos
fichero = '../out_trapezoids/output_Trapezoids_iters.csv'
data = read_csv_2(filename=fichero)

# Crear DataFrames con los resultados
df_resultados_trap = pd.DataFrame(data, columns=['nodes', 'cpus', 'time', 'err', 'qual','time_dt','err_dt','qual_dt','iters'])

# Parmetros fijos
fichero = '../out_montecarlo/output_Montecarlo_iters.csv'
data = read_csv_2(filename=fichero)

# Crear DataFrames con los resultados
df_resultados_mont = pd.DataFrame(data, columns=['nodes', 'cpus', 'time', 'err', 'qual','time_dt','err_dt','qual_dt','iters'])


# Imprimir las tablas de resultados
print("\nTabla de Resultados Trapezoides:")
print(df_resultados_trap.to_string(index=False))


# Imprimir las tablas de resultados
print("\nTabla de Resultados MonteCarlo:")
print(df_resultados_mont.to_string(index=False))


# Grfica Calidad vs Iters para Montecarlo
plt.figure(figsize=(8, 6))
plt.plot(df_resultados_mont['iters'], df_resultados_mont['qual'],label='Montecarlo', marker='o')
plt.xlabel('Iteraciones')
plt.ylabel('Calidad')
plt.xscale('log')  # Escala logartmica en el eje x
plt.title('Calidad vs Iters para Montecarlo')
plt.legend()
plt.show()

# Grfica Calidad vs Iters para Trapezoids
plt.figure(figsize=(8, 6))
plt.plot(df_resultados_trap['iters'], df_resultados_trap['qual'], color='orange',label='Trapezoides', marker='o')
plt.xlabel('Iteraciones')
plt.ylabel('Calidad')
plt.xscale('log')  # Escala logartmica en el eje x
plt.title('Calidad vs Iters para Trapezoids')
plt.legend()
plt.show()

# Grfica Calidad vs Iters comparando Montecarlo y Trapezoids
plt.figure(figsize=(8, 6))
plt.plot(df_resultados_mont['iters'], df_resultados_mont['qual'],label='Montecarlo', marker='o')
plt.plot(df_resultados_trap['iters'], df_resultados_trap['qual'],label='Trapezoides', color='orange', marker='o')
plt.xlabel('Iteraciones')
plt.ylabel('Calidad')
plt.xscale('log')  # Escala logartmica en el eje x
plt.title('Calidad vs Iters comparando Montecarlo y Trapezoids')
plt.legend()
plt.show()
```
